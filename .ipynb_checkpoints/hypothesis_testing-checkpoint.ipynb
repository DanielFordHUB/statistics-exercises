### Ace Realty wants to determine whether the average time it takes to sell homes is different for its two offices. A sample of 40 sales from office #1 revealed a mean of 90 days and a standard deviation of 15 days. A sample of 50 sales from office #2 revealed a mean of 100 days and a standard deviation of 20 days. Use a .05 level of significance.

```python
# H0 Avg TTSell @ office 1 == AVG TTSell @ office 2
# Ha Avg TTSell @ office 1 != AVG TTSell @ office 2
```

```python
alpha = 0.05
```

```python
# vissualize distributions

x = np.arange(50, 150)

y1 = stats.norm(90, 15).pdf(x)
y2 = stats.norm(100, 20).pdf(x)

plt.plot(x, y1, label = 'office 1')
plt.plot(x, y2, label = 'office 2')
plt.axvline(90, ls = ':')
plt.axvline(100, ls = ':', color = 'orange')
plt.legend()
```

```python
t, p = stats.ttest_ind_from_stats(90,15,40,100,20,50, equal_var = False)
t, p
```

```python
if p < alpha:
    print('We reject null hypothesis.')
else:
    print('We fail to reject null hypothesis.')
```

```python
mpg = data('mpg')
```

```python
mpg.head()
```

```python
# H_0: There is no difference in fuel-efficiency in cars from 2008 and 1999
# H_a: There is a difference in fuel-efficiency in cars from 2008 and 1999
```

```python
mpg['avg_fe'] = stats.hmean(mpg[['cty', 'hwy']], axis = 1)
mpg.head()
```

```python
fe_2008 = mpg[mpg.year ==2008].avg_fe
fe_1999 = mpg[mpg.year ==1999].avg_fe
```

```python
fe_2008.hist()
```

```python
fe_1999.hist()
```

```python
fe_2008.count(), fe_1999.count()
```

```python
fe_2008.var(), fe_1999.var()
```

```python
stats.levene(fe_2008, fe_1999)
```

```python
t, p = stats.ttest_ind(fe_2008, fe_1999)
t, p
```

```python
if p<alpha:
    print('reject null')
else:
    print('fail to reject null')
```

```python
fe_2008.mean(), fe_1999.mean()
```




Use the employees database.

Is there a relationship between how long an employee has been with the company and their salary?

Is there a relationship between how long an employee has been with the company and the number of titles they have had?




```python
from env import host, username, password
```

```python
def get_db_url(db):
    return f'mysql+pymysql://{username}:{password}@{host}/{db}'
```

```python
query = '''
SELECT
    DATEDIFF((SELECT MAX(hire_date)FROM employees), e.hire_date) AS tenure,
    s.salary AS salary
FROM employees e
JOIN salaries s using (emp_no)
WHERE s.to_date > NOW()
               '''
```

```python
url = f'mysql+pymysql://{username}:{password}@{host}/employees'
```

```python
df = pd.read_sql(query, url)
```

```python
df
```

### Is there a relationship between how long an employee has been with the company and their salary?


H
0
: There is no linear relationship between employee tenure and their salary


H
a
: There is a linear relationship between employee tenure and their salary

```python
stats.pearsonr(df.tenure, df.salary)
```

```python
sns.relplot(data=df, y='salary', x='tenure')
```

Is there a relationship between how long an employee has been with the company and the number of titles they have had?

```python
query = '''
SELECT
    e.emp_no,
    DATEDIFF((SELECT MAX(hire_date) FROM employees), e.hire_date) AS tenure,
    n_titles
FROM employees e
JOIN (
    SELECT emp_no, COUNT(*) as n_titles
    FROM employees e
    JOIN titles t USING (emp_no)
    GROUP BY emp_no
) emp_title_count USING (emp_no)
'''
```

```python
df2 = pd.read_sql(query, url)
```

```python
df2.shape
```

```python
df2.head()
```


H
0
: There's no linear relationship between employee tenure and number of titles

H
a
: There's a linear relationship between employee tenure and number of titles

```python
stats.pearsonr(df2.tenure, df2.n_titles)
# reject null hypothesis
```
